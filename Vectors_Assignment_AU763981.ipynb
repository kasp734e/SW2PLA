### 4. Define a function to plot the vectors of a basis of  $\mathbb{R}^3$


```python
def plotBasisR3():
    basis_vectors = np.eye(3)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.quiver(0, 0, 0, basis_vectors[0,0], basis_vectors[0,1], basis_vectors[0,2], color='r')
    ax.quiver(0, 0, 0, basis_vectors[1,0], basis_vectors[1,1], basis_vectors[1,2], color='g')
    ax.quiver(0, 0, 0, basis_vectors[2,0], basis_vectors[2,1], basis_vectors[2,2], color='b')

    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_zlim([0, 1])

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
plotBasisR3()
```
